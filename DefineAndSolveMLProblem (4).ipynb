### b. Model Building and Evaluation

```python
#prepare data

features= list(df.drop(columns = ['happiness_score','life_ladder','country'], axis=1))
y = df['happiness_score'] 
X = df[features]

print("Number of examples: " + str(X.shape[0]))
print("\nNumber of Features:" + str(X.shape[1]))
print(str(list(X.columns)))
```

```python
#fit your model to the trainign set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
#evaluate your model


lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
```

```python
y_lr_pred = lr_model.predict(X_test)
```

```python
print('Model Summary:\n')

# Print intercept (alpha)
print('Intercept:')
print('alpha = ' , lr_model.intercept_)

# Print weights
print('\nWeights:')
i = 0
for w in lr_model.coef_:
    print('w_',i+1,'= ', w, ' [ weight of ', features[i],']')
    i += 1
```

```python
# 1. Compute the RMSE using mean_squared_error()
lr_rmse = mean_squared_error(y_test, y_lr_pred, squared=False)

# 2. Compute the R2 score using r2_score()
lr_r2 = r2_score(y_test, y_lr_pred )

print('[LR] Root Mean Squared Error: {0}'.format(lr_rmse))
print('[LR] R2: {0}'.format(lr_r2))
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {'max_depth' : [4,8] , 'min_samples_leaf' : [25, 50] }
```

```python
print('Running Grid Search...')

# 1. Create a DecisionTreeRegressor model object without supplying arguments. 
#    Save the model object to the variable 'dt_regressor'

dt_regressor = DecisionTreeRegressor()


# 2. Run a Grid Search with 3-fold cross-validation and assign the output to the object 'dt_grid'.
#    * Pass the model and the parameter grid to GridSearchCV()
#    * Set the number of folds to 3
#    * Specify the scoring method

dt_grid = GridSearchCV(dt_regressor, param_grid, cv=3)


# 3. Fit the model (use the 'grid' variable) on the training data and assign the fitted model to the 
#    variable 'dt_grid_search'

dt_grid_search = dt_grid.fit(X_train, y_train)

print('Done')
```

```python
rmse_DT = -1 * dt_grid_search.best_score_
print("[DT] RMSE for the best model is : {:.2f}".format(rmse_DT) )
```

```python
dt_best_params = dt_grid_search.best_params_

dt_best_params
```

```python
dt_model = DecisionTreeRegressor(max_depth=8, min_samples_leaf=25)
dt_model.fit(X_train, y_train)
```

```python
# 1. Use the fitted model to make predictions on the test data
y_dt_pred = dt_model.predict(X_test)

# 2. Compute the RMSE using mean_squared_error()
dt_rmse = mean_squared_error(y_test, y_dt_pred, squared=False)

# 3. Compute the R2 score using r2_score()
dt_r2 = r2_score(y_test, y_dt_pred )


print('[DT] Root Mean Squared Error: {0}'.format(dt_rmse))
print('[DT] R2: {0}'.format(dt_r2))
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
estimators = [("DT", DecisionTreeRegressor(max_depth=8, min_samples_leaf=25)),
              ("LR", LinearRegression())
             ]
```

```python
print('Implement Stacking...')

stacking_model = StackingRegressor(estimators = estimators, passthrough=False)
stacking_model.fit(X_train, y_train)
print('End')
```

```python
# 1. Use the fitted model to make predictions on the test data
stacking_pred = stacking_model.predict(X_test)


# 2. Compute the RMSE 
stack_rmse = mean_squared_error(y_test, stacking_pred, squared=False)


# 3. Compute the R2 score
stack_r2 = r2_score(y_test, stacking_pred)

   
print('[Stacking] Root Mean Squared Error: {0}'.format(stack_rmse))
print('[Stacking] R2: {0}'.format(stack_r2)) 
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
print('Begin GBDT Implementation...')

gbdt_model = GradientBoostingRegressor(max_depth=2, n_estimators = 300)
gbdt_model.fit(X_train, y_train)

print('End')
```

```python
# 1. Use the fitted model to make predictions on the test data
y_gbdt_pred = gbdt_model.predict(X_test)

# 2. Compute the RMSE 
gbdt_rmse = mean_squared_error(y_test, y_gbdt_pred, squared=False)


# 3. Compute the R2 score 
gbdt_r2 = r2_score(y_test, y_gbdt_pred)


print('[GBDT] Root Mean Squared Error: {0}'.format(gbdt_rmse))
print('[GBDT] R2: {0}'.format(gbdt_r2))  
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
print('Begin RF Implementation...')

rf_model = RandomForestRegressor(max_depth=32, n_estimators = 300)
rf_model.fit(X_train, y_train)

print('End')
```

```python
# 1. Use the fitted model to make predictions on the test data
y_rf_pred = rf_model.predict(X_test)


# 2. Compute the RMSE 
rf_rmse = mean_squared_error(y_test, y_rf_pred, squared=False)

# 3. Compute the R2 score 
rf_r2 = r2_score(y_test, y_rf_pred)


print('[RF] Root Mean Squared Error: {0}'.format(rf_rmse))
print('[RF] R2: {0}'.format(rf_r2))  
```

```python
RMSE_Results = [stack_rmse, lr_rmse, dt_rmse, gbdt_rmse, rf_rmse]
R2_Results = [stack_r2, lr_r2, dt_r2, gbdt_r2, rf_r2]

labels = ['Stacking','LR', 'DT', 'GBDT', 'RF']
rg= np.arange(len(labels))
width = 0.35


# 1. Create bar plot with RMSE results
plt.bar(rg - width/2, RMSE_Results, width=width, label='RMSE')

# 2. Create bar plot with R2 results
plt.bar(rg + width/2, R2_Results, width=width, label='R2')



plt.xticks(rg, labels)

plt.xlabel("Models")
plt.ylabel("Score")


plt.ylim([0,1.1])
plt.title('Model Performance Comparision')
plt.legend()
plt.show()

```

### Analysis

The range of R^2 scores of ensemble models like stacking, GBDT, and RF, is higher (higher range) compared to the individual models like DT and LR. The range of RSME scores of ensemble models is lower (smaller range) compared to individual models. From the above bar plot, the Gradient-Boosted Decision Tree (GBDT) is the best performing model which Im going to move forward with. The other two ensemble models are also performing better compared to individual models, especially the Random Forest model. High R^2 and low RSME gives the best performing model. So, from the above graph the range of R^2 score of GBDT is high and the range of RSME score is low.
```python
#1. Obtain "feature importance" scores from the model object and save the array to the variable 
# 'feature_imp'

feature_imp = gbdt_model.feature_importances_

#2. Create a Pandas DataFrame with a list of all features and their scores. 
# Save the result to the variable 'df_features'
df_features = pd.DataFrame({'name': X_train.columns.values, 'imp': feature_imp})


#3. Sort df_features in descending order and
# save the result to the variable 'df_sorted'
df_sorted = df_features.sort_values(by ='imp',ascending =False)


#4. Obtain the top 5 sorted feature names and save the result to list 'top_five' 
top_five = list(df_sorted.iloc[:5]['name'])

print('Top five features: {0}'.format(top_five))
```
